```python
#!/usr/bin/env python3

import itertools
import scipy
import scikits.bootstrap as bootstrap
import pandas as pd
import numpy as np
import os.path as osp
import os
```

```python

def my_agg(x):
    names = {'mean':np.mean(x,out=None),
         'n':len(x),
         'std':np.std(x,out=None),
         'cimin':bootstrap.ci(x, statfunction=scipy.mean)[0],
         'cimax':bootstrap.ci(x, statfunction=scipy.mean)[1]}
    return pd.Series(names,index= ['mean', 'n','std', 'cimin',
                                   'cimax']).astype(float)
results=[]
basedir='/Volumes/3d_ltmp/exports/structural_complexity/OM/'
for f in os.listdir(basedir):
    if f.endswith(".txt"):
        df=pd.read_csv(os.path.join(basedir,f))
        df=df.dropna()
        df=pd.melt(df, 
          id_vars=['camp','reefname','site','transect','qsize'],
           value_vars=['rgsty','slope','aspect','rangez','sdevz','rgstyXY','concavity','meandevz'],
           var_name='metric',value_name='value')

        df=df.groupby(['camp','reefname',
               'site','transect',
               'qsize','metric']).apply(my_agg).reset_index()
        df=pd.DataFrame(df)
        
    results.append(df)


results=pd.concat(results, axis=0)
```
